```python
import pandas as pd

remuneracao_servidores = 'https://raw.githubusercontent.com/DeboraGomes23/IESB_trabalho/main/remuneracaoServidores.csv'

df = pd.read_csv(remuneracao_servidores,sep=';')

print(df.columns.tolist())
```

```python
import pandas as pd

print(df.dtypes)
```

#Qual o espaço amostral compreendido nesta base de dados?

```python
valores_unicos_concatenados = df.melt().drop_duplicates()['value']
espaco_amostral_total = valores_unicos_concatenados.nunique()

espaco_amostral_total = '{:,}'.format(espaco_amostral_total)

print("Espaço amostral total em todas as colunas:", espaco_amostral_total)
```

# Quantos servidores estão lotados no corpo de bombeiros?

```python
servidores_bombeiros = df[df['ÓRGÃO'] == 'CORPO DE BOMBEIRO MILITAR DO DISTRITO FEDERAL - SIAPE']
num_servidores_bombeiros = servidores_bombeiros.shape[0]
num_servidores_bombeiros_formatado = '{:,}'.format(num_servidores_bombeiros)

print("O número de servidores lotados no Corpo de Bombeiros é:", num_servidores_bombeiros_formatado)
```

# Qual o órgão público com maior número de funcionários?

```python
num_funcionarios_por_orgao = df['ÓRGÃO'].value_counts()
orgao_maior_numero_funcionarios = num_funcionarios_por_orgao.idxmax()
maior_numero_funcionarios = num_funcionarios_por_orgao.max()
maior_numero_funcionarios_formatado = '{:,}'.format(maior_numero_funcionarios)

print("O órgão público com o maior número de funcionários é:", orgao_maior_numero_funcionarios)
print("Número de funcionários:", maior_numero_funcionarios_formatado)
```

# Qual a probabilidade de ao escolher uma linha ao acaso, ser de um funcionário que trabalha na secretaria de educação?

```python
num_funcionarios_secretaria_educacao = df[df['ÓRGÃO'] == 'SECRETARIA DE ESTADO DE EDUCACAO                                                '].shape[0]
num_total_funcionarios = df.shape[0]

probabilidade_secretaria_educacao = num_funcionarios_secretaria_educacao / num_total_funcionarios * 100
probabilidade_secretaria_educacao_arredondada = round(probabilidade_secretaria_educacao, 2)

print("A probabilidade de escolher uma linha ao acaso e ela ser de um funcionário que trabalha na Secretaria de Educação é:", probabilidade_secretaria_educacao_arredondada, "%")
```

# Qual órgão publico possui a maior média salarial?

```python
df['BRUTO'] = df['BRUTO'].str.replace(',', '.').astype(float)

media_salarial_por_orgao = df.groupby('ÓRGÃO')['BRUTO'].mean()

orgao_maior_media_salarial = media_salarial_por_orgao.idxmax()
maior_media_salarial = media_salarial_por_orgao.max()

maior_media_salarial_formatada = '{:,.2f}'.format(maior_media_salarial)

print("O órgão público com a maior média salarial é:", orgao_maior_media_salarial)
print("Média salarial:", maior_media_salarial_formatada)
```

# O servidor com maior remuneração básica pertence a qual órgão?

```python
from babel.numbers import format_currency

df['REMUNERAÇÃO BÁSICA'] = df['REMUNERAÇÃO BÁSICA'].str.replace(',', '.').astype(float)

df_ordenado = df.sort_values(by='REMUNERAÇÃO BÁSICA', ascending=False)

orgao_servidor_maior_remuneracao = df_ordenado.iloc[0]['ÓRGÃO']
maior_remuneracao_basica = df_ordenado.iloc[0]['REMUNERAÇÃO BÁSICA']
maior_remuneracao_formatada = format_currency(maior_remuneracao_basica, 'BRL', locale='pt_BR')

print("O servidor com a maior remuneração básica pertence ao órgão:", orgao_servidor_maior_remuneracao)
print("Maior remuneração básica:", maior_remuneracao_formatada)
```

# Qual o valor pago para todos os funcionários públicos?

```python
df['BRUTO'] = df['BRUTO'].astype(float)
valor_total_pago = df['BRUTO'].sum()

valor_formatado = "R$ {:,.2f}".format(valor_total_pago)

print("O valor total pago para todos os funcionários públicos é:", valor_formatado)
```

# Elabore um gráfico de pizza com a remuneração bruta dos servidores por

```python
unique_df = df.drop_duplicates()
grouped_data = unique_df.groupby('ÓRGÃO')['BRUTO'].sum()
top_10_orgaos = grouped_data.nlargest(100)
plt.figure(figsize=(5, 5))
plt.pie(top_10_orgaos, labels=top_10_orgaos.index, autopct='%1.1f%%', startangle=140)
plt.title('Remuneração Bruta dos Servidores por Orgão')
plt.axis('equal')
plt.show()
```

# Identifique qual órgão possui salário liquido com menos variações, utilizando
# para isto medidas de dispersão:

```python
df['LÍQUIDO'] = df['LÍQUIDO'].replace(',', '.').astype(float)
desvio_padrao_por_orgao = df.groupby('ÓRGÃO')['LÍQUIDO'].std()

orgao_menos_variacoes = desvio_padrao_por_orgao.idxmin()
menor_desvio_padrao = desvio_padrao_por_orgao.min()

print("O órgão com o salário líquido com menos variações é:", orgao_menos_variacoes)
print("Desvio padrão:", menor_desvio_padrao)
```

# Calcule o índice de correlação entre o IRRF e os salários liquido e bruto. Em
# qual dos casos o índice de correlação foi maior?

```python
df['IRRF'] = df['IRRF'].str.replace(',', '.').astype(float)
df['LÍQUIDO'] = df['LÍQUIDO'].astype(str).str.replace(',', '.').astype(float)

correlation_irrf_liquido = df['IRRF'].corr(df['LÍQUIDO'])
correlation_irrf_bruto = df['IRRF'].corr(df['BRUTO'])

if abs(correlation_irrf_liquido) > abs(correlation_irrf_bruto):
    maior_correlacao = 'LÍQUIDO'
else:
    maior_correlacao = 'BRUTO'

print("Coeficiente de correlação entre IRRF e salário líquido:", correlation_irrf_liquido)
print("Coeficiente de correlação entre IRRF e salário bruto:", correlation_irrf_bruto)
print("A maior correlação foi encontrada com o salário: ", maior_correlacao)
```

# Adicione uma nova coluna que irá conter a diferença entre o salario bruto e
# liquido, e responda as seguintes questões:

```python
df['diferenca_salario'] = df['BRUTO'] - df['LÍQUIDO']

print(df.head())
```

# Qual a correlação entre o IRRF pago e esta nova coluna?

```python
correlacao = df['IRRF'].corr(df['diferenca_salario'])

correlacao_2 = '{:.2f}'.format(correlacao)

print("Correlação entre o IRRF pago e a diferença entre salário bruto e líquido:", correlacao_2)
```

# Qual órgão apresenta índice de correlação entre IRRF e diferença
# salarial maior?

```python
df['diferenca_salario'] = df['BRUTO'] - df['LÍQUIDO']

correlacoes_por_orgao = df.groupby('ÓRGÃO')[['IRRF', 'diferenca_salario']].corr().iloc[0::2,-1]

orgao_maior_correlacao = correlacoes_por_orgao.groupby('ÓRGÃO').mean().idxmax()

print("Órgão com maior correlação entre IRRF e com diferença salarial maior é:", orgao_maior_correlacao)
```
